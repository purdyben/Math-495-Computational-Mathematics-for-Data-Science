```python
# set up
titanic = pd.read_csv("titanic.csv") 
titanic = titanic.rename({'home.dest': 'dest'},axis='columns')
def split(word): 
    return [char for char in word]
alphabet = set("abcdefghijklmnopqrstuwxyz")
```

```python
print("problem 1: ", '\n')
partA = split("pneumonoultramicroscopicsilicovolcanoconiosisis")
partB = set(sorted(partA))
partC = alphabet - partB
partD = dict.fromkeys(partB,0);
for ch in partA:
    partD[ch] = partD[ch] + 1;
    
print("part A :", partA, '\n')
print("part B : ",partB, '\n')
print("part C : ", partC, '\n')
print("Dictionary D", partD, '\n')
```

```python
print("problem 2: ")
twoA = set("supercalifragilisticexpialidocious")

unionAB = sorted(partB.union(twoA))
print("part A : A U B :",unionAB, '\n')
AnotB = (twoA - partB)
BnotA = (partB - twoA)
print("A - B :", AnotB)
print("B - A :", BnotA)

AdeltaB = AnotB.union(BnotA)
print("A ∆ B :", AdeltaB)

A = twoA
B = partB
print("A - B = A intersection B complement")
if A - B == A.intersection(alphabet - B):
    print(True)
else:
    print(False)

print("A - B = A complement intsection B")
if A - B == (alphabet - A).intersection( B):
    print(True)
else:
    print(False)
    
print("\nDe Morgan’s law \n")
# Definition of De Morgan's law: ... The complement of the 
# union of two sets is equal to the intersection of their 
# complements and the complement of the intersection of two sets
# is equal to the union of their complements. These are called De Morgan's laws.
print("(A U B)' = A' ∩ B'")
print("(A U B)' = ", alphabet - B.union(A))
print(" A' ∩ B' = ",(alphabet - B).intersection(alphabet - A), ": True" )
print("(A U B)' = A' ∩ B'  : True\n" )
print("(A ∩ B)' = A' U B'")
print("(A ∩ B)' = ", (alphabet - B.intersection(A)))
print(" A' U B' = ", (alphabet - B).union(alphabet - A), ": True")




```

```python
print("problem 4:")

fareMean = (titanic['fare']).mean()
subsetFM = titanic.query('fare-@fareMean < .5')
# print(len(subsetFM))
print("part A : number of rows: ", len(subsetFM))
print(subsetFM.iloc[:3])
def FindNumberOfCommonStrings():
    ticket = titanic['ticket'].dropna().sort_values(ascending= True, na_position='first')
    prev = -1 
    curr = -1
    seen = 0
    count = 0
    for index in ticket.index:
        curr = ticket.loc[index]
        if prev != -1:
            if(prev == curr and seen != 1):
                seen = 1;
                count = count + 1
            else:
                seen = 0
        prev = curr 
    return count
    
print("part B : ", FindNumberOfCommonStrings())

def setUpLucky():
    l = []
    for index in titanic.index:
        row = titanic.loc[index]
        if row['pclass'] == 1 and row['fare'] > fareMean and row['survived'] == 0:
            l.append('unlucky')
        elif row['pclass'] > 1 and row['fare'] <= fareMean and row['survived'] == 1 :
            l.append('lucky')
        else:

         l.append('averageluck')
    return l
titanic = titanic.assign(luck = setUpLucky()) 
print("part C : \n", titanic['luck'].iloc[:4])

print("part D : ", titanic.groupby('luck').size())
```

```python
print("problem 2: ")
print("Part A: ")
DF1 = {'id' : ['s01','s05','s06','s08'],
       'Name' : ['Julie','Carl','Luke','David'],
       'City' : ['Des Moines','Iowa city','Chicago','Minneapolis'],
       'Age' : ['25','34','29','45']}
DF2 = {'id' : ['s01','s08','s03','s04'],
       'Name' : ['Julie','David','Bradley','Nina'],
       'City' : ['Des Moines','Minneapolis','Tempe','Pittsburg'],
       'Age' : [3.25,3.4,3.3,4]}


df1 = pd.DataFrame(DF1, columns = ['id', 'Name', 'City','Age'])
df2 = pd.DataFrame(DF2, columns = ['id', 'Name', 'City','Age'])
df_left = pd.merge(df1, df2, on='id', how='left')
print(df_left,'\n')
print('Left join :  Takes each column from and labels it column_x and column_y to dictate which data frame that column came from. Then is creates a new data frame with all columns ordered x - y. Columns stay in order as the prevues data frames. By id dictates what data is merges. all of x is added but if a data frame y doesn’t have the corresponding id as x its replaced with NaN')
print("Part B: ")
df_right = pd.merge(df1, df2, on='id', how='right')
print(df_right,'\n')
print('Right join: Labels and orders all columns just as Left join but Id are now the Ids of data frame y. NaN is added to any x row that doesn’t line up with the id ')
print("Part C: ")
df_inner = pd.merge(df1, df2, on='id', how='inner')
print(df_inner,'\n')
print('Inner join is the interaction of the ids and the responding rows ')
print("Part D: ")
df_outer = pd.merge(df1, df2, on='id', how='outer')
print(df_outer,'\n')
print('Outer join is all data ids from data frame x and data frame y, or you could think of it as the union of left join and right join')
```
